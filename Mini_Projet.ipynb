```python
moyenne_bac = float(input("Quel est votre moyenne au bac? "))

if 12 <= moyenne_bac < 14:
   print("Assez bien")
elif 14 <= moyenne_bac < 16:
   print("Bien")
elif 16 <= moyenne_bac < 18:
   print("Très bien")
elif moyenne_bac >= 18:
   print("Les félicitation du jury")
else:
   print("Pas de mention")
```

```python
nb_de_vies = 8

mot_mystère = "violon"
mot_correct = "_" * len(mot_mystère)

while nb_de_vies > 0 and mot_mystère != mot_correct:
   lettre = input("Entrez une lettre: ")

   if lettre in mot_mystère:
      for i in range(len(mot_mystère)):
         if mot_mystère[i] == lettre:
           mot_correct = mot_correct[:i] + lettre + mot_correct[i + 1:]
      print("Bien joué ! Le mot est", mot_correct)
   else:
      nb_de_vies -= 1

   if nb_de_vies == mot_correct:
      print("Bien joué ! Le mot est", mot_correct)
   elif nb_de_vies == 0:
      print("Vous avez perdu !")
   else:
      print("Il vous reste", nb_de_vies)
      print("Felicitation le mot secret est bien :", mot_correct)
```
